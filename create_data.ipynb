```python
from os import listdir
from PIL import Image as PImage
import os
import glob
```

```python
files = glob.glob('./images/cropped/*')
for f in files:
    os.remove(f)
```

```python
newsize = (300, 300) 

path = "./images/base/"

def loadImages(path):
    # return array of images

    imagesList = listdir(path)
    loadedImages = []
    for image in imagesList:
        img = PImage.open(path + image)
        img = img.resize(newsize)
        loadedImages.append(img)
        #break
    return loadedImages

# your images in an array
imgs = loadImages(path)

#def transpose_save(img):

img_cntr = 0   

for img in imgs:  
    #Split 1
    img1 = img.crop((0,0,150,150))
    img_cntr = img_cntr + 1
    img1.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img = img1.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")

    #Split 2
    img2 = img.crop((150,0,300,150))
    img_cntr = img_cntr + 1
    img2.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img = img2.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    #Split 3
    img3 = img.crop((0,150,150,300))
    img_cntr = img_cntr + 1
    img3.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img = img3.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")

    #Split 4
    img4 = img.crop((150,150,300,300))
    img_cntr = img_cntr + 1
    img4.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img = img4.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")
    
    img_cntr = img_cntr + 1
    transposed_img =  transposed_img.rotate(90)
    transposed_img.save("./images/cropped/" + str(img_cntr) + ".jpg")

print(img_cntr)
```

```python
import random
a = list(range(224))
a = [x+1 for x in a]
random.shuffle(a)
train_cnt = a[0:155]
test_cnt = a[155:]

files = glob.glob('./images/train/*')
for f in files:
    os.remove(f)
    
files = glob.glob('./images/test/*')
for f in files:
    os.remove(f)

import shutil, os
for x in train_cnt:
    #print(x)
    fname = './images/cropped/' + str(x) + ".jpg"
    shutil.copy(fname, './images/train')
    
for x in test_cnt:
    #print(x)
    fname = './images/cropped/' + str(x) + ".jpg"
    shutil.copy(fname, './images/test') 
```
