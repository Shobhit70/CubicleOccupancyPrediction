```python
train_files = glob.glob('./images/train/*.jpg')
test_files = glob.glob('./images/test/*.jpg')
#print(len(train_files), len(test_files))
#Segregate and move train data

#print(len(train_files))
for n in train_files:
    f_name = n.split('\\')[1]

    if f_name in list(ground_truth['Image Name']):
        val = ground_truth.index[ground_truth['Image Name'] == f_name]   
                
        if ground_truth.iloc[val[0]][1] == 0:
            shutil.copy('./images/train/' + f_name, './images/train/empty/')
        else:
            shutil.copy('./images/train/' + f_name, './images/train/occupied/')

#Segregate and move test data
c=0
for n in test_files:
    f_name = n.split('\\')[1]
    
    if f_name in list(ground_truth['Image Name']):
        val = ground_truth.index[ground_truth['Image Name'] == f_name]   
                
        if ground_truth.iloc[val[0]][1] == 0:
            shutil.copy('./images/test/' + f_name, './images/test/empty/')
        else:
            shutil.copy('./images/test/' + f_name, './images/test/occupied/')
```
